```python
import pysam
```

```python
file = pysam.AlignmentFile("/sbgenomics/project-files/merged-tumor.bam")
```

```python
cnt_unmapped_reads = 0
cnt_total_reads = 0
cnt_mapping_quality_0 = 0
mapping_quality_sum = 0

for read in file:
    if read.is_unmapped:
        cnt_unmapped_reads += 1
    cnt_total_reads += 1
    if read.mapping_quality == 0:
        cnt_mapping_quality_0 += 1
    mapping_quality_sum += read.mapping_quality

average_mapping_quality = mapping_quality_sum / cnt_total_reads
average_mapping_quality_without_0 = mapping_quality_sum / (cnt_total_reads - cnt_mapping_quality_0)

print(f"Number of unmapped reads: {cnt_unmapped_reads}")
print(f"Total number of reads: {cnt_total_reads}")
print(f"Number of reads with mapping quality 0: {cnt_mapping_quality_0}")
print(f"Average mapping quality for all the reads: {average_mapping_quality}")
print(f"Average mapping quality if reads with 0 mapp quality are filtered out: {average_mapping_quality_without_0}")
```
