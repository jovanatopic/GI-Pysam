```python
import pysam
```

```python
file = pysam.AlignmentFile("/sbgenomics/project-files/merged-tumor.bam")
```

```python
first = 0
cnt_unmapped_reads = 0
cnt_total_reads = 0
cnt_mapping_quality_0 = 0
mapping_quality_sum = 0

for read in file:
    if first == 0:
        first_read = read
        first = -1
    if read.is_unmapped:
        cnt_unmapped_reads += 1
    cnt_total_reads += 1
    if read.mapping_quality == 0:
        cnt_mapping_quality_0 += 1
    mapping_quality_sum += read.mapping_quality

average_mapping_quality = mapping_quality_sum / cnt_total_reads
average_mapping_quality_without_0 = mapping_quality_sum / (cnt_total_reads - cnt_mapping_quality_0)

print(f"The first read from the file:\n {first_read}")
print(f"The flag field of the first read: {first_read.flag}")
print(f"The number of unmapped reads: {cnt_unmapped_reads}")
print(f"The total number of reads: {cnt_total_reads}")
print(f"The number of reads with mapping quality 0: {cnt_mapping_quality_0}")
print(f"The average mapping quality for all the reads: {average_mapping_quality}")
print(f"The average mapping quality if reads with 0 mapp quality are filtered out: {average_mapping_quality_without_0}")
```
